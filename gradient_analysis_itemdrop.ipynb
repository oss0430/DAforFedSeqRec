```python
import ast
local_results = log_content.split('(client:354)')[1:]
refined_local_results = []  

for local_result in local_results :
    refined_str = local_result.split('\n')[0]
    refined_str = refined_str.split('INFO: ')[1]
    refined_dict = ast.literal_eval(refined_str)
    refined_local_results.append(refined_dict)
    
print(refined_local_results[0]['Round'])
print(refined_local_results[-1])
```

```python
## segregate attacker and non-attacker clients

normal_clients_result = []
dropped_clients_result = []

def parse_client_id(role : str) -> int :
    """
    Parse the client id from the role string such as 'Client #593'.
    """
    return int(role.split('#')[1])

for local_result in refined_local_results:
    current_id = parse_client_id(local_result['Role'])
    
    if current_id in federated_configuration["dropping_user_id"]:
        dropped_clients_result.append(local_result)
    else:
        normal_clients_result.append(local_result)

print(len(dropped_clients_result))
print(len(normal_clients_result))
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

## x-axis : round, y-axis : gradient mean, std, max, min

## for each round gradient values are
normal_clients_gradients = {i : [] for i in range(federated_configuration["total_round_num"])}
dropped_clients_gradients = {i : [] for i in range(federated_configuration["total_round_num"])}

for local_result in normal_clients_result:
    current_round = local_result['Round']
    if current_round > federated_configuration["total_round_num"]-1:
        break
    normal_clients_gradients[current_round].append(local_result['Results_raw']['train_avg_loss'])
    
for local_result in dropped_clients_result:
    current_round = local_result['Round']
    if current_round > federated_configuration["total_round_num"]-1:
        break
    dropped_clients_gradients[current_round].append(local_result['Results_raw']['train_avg_loss'])

benign_datas = [value for key, value in normal_clients_gradients.items()]

fig, ax = plt.subplots()

benign_boxplot = ax.boxplot(benign_datas, showfliers=False, patch_artist=True, vert=True, labels=normal_clients_gradients.keys(), manage_ticks=False)


mean_dropped_clients_gradients = np.zeros((federated_configuration["total_round_num"]))
for round, gradients in dropped_clients_gradients.items():
    if len(gradients) == 0:
        mean_dropped_clients_gradients[round] = np.nan
    else: 
        mean_dropped_clients_gradients[round] = np.mean(gradients)

dropped_line_plot = ax.plot(mean_dropped_clients_gradients, label='Dropped', color='red')

plt.xlabel('Round')
ax.set_xticks(np.arange(0, federated_configuration["total_round_num"], 20))
ax.grid(True)
plt.ylabel('Mean Local Gradient')

handles, labels = ax.get_legend_handles_labels()
empty_boxplot_label = mpatches.Patch(color='black', label='Normal')
handles.extend([empty_boxplot_label]) 
ax.legend(handles=handles)

plt.show()



```
